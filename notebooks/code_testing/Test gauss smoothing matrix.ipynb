```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, "../../")
import ompy

%load_ext autoreload
%matplotlib notebook
```

```python
raw = ompy.read_mama_2D("Si28_raw_matrix_compressed.m")
raw.plot(zscale="linear")
```

```python
raw_smoothed_mat = ompy.gauss_smoothing_matrix(raw.matrix, raw.E1_array, fwhm=2000)
import copy
raw_smoothed = copy.deepcopy(raw)
raw_smoothed.matrix = raw_smoothed_mat
raw_smoothed.plot(zscale="linear")
plt.show()
```

```python
%matplotlib notebook
f, ax = plt.subplots(1)

print("raw.calibration() =", raw.calibration())

#E_limits = [2000, 2050]
#raw.plot_projection_x(E_limits, ax=ax, label="raw")
#raw_smoothed.plot_projection_x(E_limits, ax=ax, label="smoothed")

i_Ex = 80
ax.plot(raw.E1_array, raw.matrix[i_Ex, :], label="raw")
ax.plot(raw_smoothed.E1_array, raw_smoothed.matrix[i_Ex, :], label="smoothed")


ax.legend()

plt.show()
```

```python
E_array = np.linspace(0, 10000, 200)
delta_energy = E_array[1] - E_array[0]

gauss1 = ompy.gaussian(E_array, mu=500, sigma=50)
print("np.sum(gauss1)*delta_energy =", np.sum(gauss1)*delta_energy)

gauss2 = ompy.gaussian(E_array, mu=500, sigma=6.8/2.355)
print("np.sum(gauss2)*delta_energy =", np.sum(gauss2)*delta_energy)

f, ax = plt.subplots(1,1)
ax.plot(E_array, gauss1)
ax.plot(E_array, gauss2)

plt.show()
```

```python
i_Ex = 70

E_array = raw.E1_array
raw_row = raw.matrix[i_Ex, :]
raw_row_smoothed = ompy.gauss_smoothing(raw_row, E_array, fwhm=678)

print("np.sum(raw_row) =", np.sum(raw_row))
print("np.sum(raw_row_smoothed) =", np.sum(raw_row_smoothed))
print("ratio =", np.sum(raw_row_smoothed)/np.sum(raw_row))

f, ax = plt.subplots(1)

ax.plot(E_array, raw_row, label="raw")
ax.plot(E_array, raw_row_smoothed, label="smoothed")

ax.legend()
```
